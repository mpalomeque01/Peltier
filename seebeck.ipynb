```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import metodos_labo as labo
from scipy.optimize import curve_fit

plt.style.use('./informes.mplstyle')
```

I = 0.2

```python
fig, ax = plt.subplots(3,1, figsize = (10,8))

df = pd.read_csv('Mediciones/Clase 3/celda_temp_fija_I=0.2.csv')

t = df['Tiempo [s]'].values
v = df['Tension generador [V]'].values
tf = df['Tensin termocupla fria [V]'].values*1e3
tc = df['Tensin termocupla caliente [V]'].values*1e3
#Convierto voltaje a temperatura con la calibracin (no anotamos en codigo cual es cual)
tf = tf*22.63 + 28.07
tc = tc*22.81 + 27.90

delta_T = tc-tf

def ajuste_lineal(x,m,b):
    return m*x + b

popt, pcov = curve_fit(ajuste_lineal,delta_T,v)
print(f'Seebeck = {popt}')
print(f'popt_sd = {np.sqrt(np.diag(pcov))}')

fig.suptitle('I = 0.2', fontsize=16)
ax[0].plot(t,v)
ax[1].plot(t,delta_T)
ax[2].scatter(delta_T,v,marker = '.')
for i in range(3):
    ax[i].tick_params(axis='both', which='major', labelsize=10)
    ax[i].tick_params(axis='both', which='minor', labelsize=8)
ax[0].set_ylabel('Voltaje [mV]', fontsize=13)
ax[1].set_ylabel(f'$\Delta T\  [C]$', fontsize=13)
ax[1].set_xlabel(f'Tiempo [s]', fontsize=13)
ax[2].set_ylabel('Voltaje [mV]', fontsize=13)
ax[2].set_xlabel(f'$\Delta T\  [C]$', fontsize=13)
plt.subplots_adjust(hspace=0.5)
fig.savefig('Figuras/Hielo y Corriente 0.2A.pdf')
```

```python
corrientes = [0.2,0.4,0.6,0.8,1]
t = []
v = []
tf = []
tc = []
delta_T = []
for i in corrientes:
    df = pd.read_csv(f'Mediciones/Clase 3/celda_temp_fija_I={i}.csv')
    t.append(df['Tiempo [s]'].values)
    v.append(df['Tension generador [V]'].values)
    tf_temp = df['Tensin termocupla fria [V]'].values*1e3
    tc_temp = df['Tensin termocupla caliente [V]'].values*1e3
    tf_temp = tf_temp*22.63 + 28.07
    tc_temp = tc_temp*22.81 + 27.90
    delta_T.append(tc_temp-tf_temp)
    tf.append(tf_temp)
    tc.append(tc_temp)
letra = 10
fig, ax = plt.subplots(1,2,figsize = (16,5))
for i in range(len(corrientes)):
    ax[0].plot(t[i],delta_T[i],label = f'I = {corrientes[i]}')
    ax[1].scatter(delta_T[i],v[i],label = f'I = {corrientes[i]}')
ax[0].set_xlabel('Tiempo [s]',fontsize = letra)
ax[0].set_ylabel('$\Delta T$',fontsize = letra)
ax[0].set_title('Agua con hielo y resistencia')
ax[0].legend(fontsize = 'medium')
ax[1].set_xlabel('$\Delta T$',fontsize = letra)
ax[1].set_ylabel('$Voltaje [mV]$',fontsize = letra)
ax[1].set_title('Coeficiente de Seebeck')
ax[1].legend(fontsize = 'medium')
for i in range(2):
    ax[i].tick_params(axis='both', which='major', labelsize=10)
    ax[i].tick_params(axis='both', which='minor', labelsize=8)
fig.savefig('Figuras/Diferencias de temperaturas seebeck.pdf')
```
